IMPORTING REQUIRED LIBRARIES AND DATASETS

```python
#Importing the respective datasets into session
import pandas as pd

Tele1 = pd.read_csv('Telcom_dataset.csv')
Tele1.head()
```

```python
#Importing the respective datasets into session
Tele2 = pd.read_csv('Telcom_dataset2.csv')
Tele2.head()
```

```python
#Importing the respective datasets into session
Tele3 = pd.read_csv('Telcom_dataset3.csv')
Tele3.head()
```

```python
#Importing the respective datasets into session
Cell_Geo = pd.read_csv('cells_geo.csv')
Cell_Geo.head()
```

CLEANING UP DATA

```python
#Renaming column names to match across the imported datasets
Tele1= Tele1.rename(columns={'PRODUTC':'PRODUCT', 'DATETIME':'DATE_TIME'})
Tele1.head()
```

```python
#Renaming column names to match across the imported datasets
Tele2= Tele2.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT', 'DW_B_NUMBER':'DW_B_NUMBER_INT'})
Tele2.head()
```

```python
#Renaming column names to match across the imported datasets
Tele3= Tele3.rename(columns={'CELLID':'CELL_ID', 'SIET_ID': 'SITE_ID'})
Tele3.head()
```

```python
#Splitting the cell_geo data into separate columns based on the ';'
Cell_Geo = Cell_Geo[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].str.split(';',n=11, expand=True)
Cell_Geo.head()
```

```python
#Renaming column names accordingly from numeric values to their respective names
Cell_Geo= Cell_Geo.rename(columns={1:'VILLES',2:'STATUS',3:'LOCALISATION',4:'DECOUPZONE',5:'ZONENAME',6:'LONGITUDE',7:'LATITUDE',8:'REGION',9:'AREA',10:'CELL_ID',11:'SITE_CODE'})
Cell_Geo.head()
```

```python
#Deleting unnecessary columns
del Cell_Geo[0]
Cell_Geo.head()
```

```python
#Merging the 3 day datasets based on matching attributes
Tele = pd.concat([Tele1,Tele2,Tele3])
Tele.head()
```

```python
#Deleting unnecessary columns
Tele.drop(columns=["COUNTRY_A", "COUNTRY_B", "CELL_ON_SITE"], inplace=True)
Tele.head()
```

```python
#Counting rows with null values
# print(Tele[Tele.isnull().any(axis=1)][null_columns].head())
null_columns=Tele.columns[Tele.isnull().any()]
Tele[null_columns].isnull().sum()
```

```python
#showing output for columns with null values
print(Tele[Tele["SITE_ID"].isnull()][null_columns])
# Tele["SITE_ID"].fillna("STH", inplace = True) 
```

```python
#Showing number of rows with duplicate values in the entire merged dataframe
Tele.duplicated().sum()
```
